# Question 4

Step 1. Import the necessary libraries

Step 2. Import the dataset from the attached file wind.txt

```python
# I was issue when importing data from link so I have converted it to csv and then imported
import pandas as pd
import datetime
data = pd.read_csv("wind.csv")
data.head()
```

Step 3. Assign it to a variable called data and replace the first 3 columns by a proper
datetime index

```python
#Joining collumn with '-' and adjusting column
New = data['Yr'].astype(str)+'/'+data['Mo'].astype(str)+'/'+data['Dy'].astype(str)
data.insert(0, 'Date', New)
data['Date'] =  pd.to_datetime(data['Date'])
#Deleting the columns that we have merged
del data['Yr']
del data['Mo']
del data['Dy']
data.head()
```

Step 4: Year 2061? Do we really have data from this year? Create a function to fix it and apply it.

```python
#creating function to fix 
def Fix(format):    
    if format.year > 2000:
        year = format.year-100
    else:
        year = format.year
    return datetime.date(year,format.month,format.day)

data['Date'] = data['Date'].apply(Fix)
data.head()
```

Step 5. Set the right dates as the index. Pay attention at the data type, it should be
datetime64[ns].

```python
data = data.set_index('Date')
data
```

Step 6. Compute how many values are missing for each location over the entire
record.They should be ignored in all calculations below.


```python
data.isna().sum()
```

Step 7. Compute how many non-missing values there are in total.

```python
data.count()
```

Step 8. Calculate the mean windspeeds of the windspeeds over all the locations and
all the times.

```python
Mean = data.mean().mean()
Mean
```

Step 9. Create a DataFrame called loc_stats and calculate the min, max and mean
windspeeds and standard deviations of the windspeeds at each location over all the
days

```python
loc_min = pd.Series(data.min(axis=0),name="Minimum")
loc_max = pd.Series(data.max(axis=0),name="Maximum")
loc_mean = pd.Series(data.mean(axis=0),name="Mean")
loc_std = pd.Series(data.std(axis=0),name="Standard Deviation")
loc_stats = pd.concat([loc_min, loc_max, loc_mean, loc_std], axis=1)
loc_stats
```

Step 10. Create a DataFrame called day_stats and calculate the min, max and mean
windspeed and standard deviations of the windspeeds across all the locations at each
day

```python
min1 = pd.Series(data.min(axis=1),name="Minimum")
max1= pd.Series(data.max(axis=1),name="Maximum")
mean1 = pd.Series(data.mean(axis=1),name="Mean")
std1 = pd.Series(data.std(axis=1),name="Standard Deviation")
day_stats = pd.concat([min1, max1, mean1, std1], axis=1)
day_stats
```

# Question 6

Create a line plot showing the number of marriages and divorces per capita in the
U.S. between 1867 and 2014. Label both lines and show the legend.

```python
import pandas as pd
US = pd.read_csv("us-marriages-divorces-1867-2014.csv")
US.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
USYear = pd.DataFrame(US,columns=['Year'])
US_Data = pd.DataFrame(US,columns=['Marriages_per_1000','Divorces_per_1000']) 
fig, ax = plt.subplots(figsize=(8, 8))
for column in US_Data:
    ax.plot(US['Year'], US_Data[column])
ax.set_title('No. of Marriages and Divorces per Capita in the U.S. between 1867 and 2014')
ax.legend(US_Data)
plt.xlabel("Year")
plt.ylabel("Per Capita")
plt.show()
```

# Question 7

Create a vertical bar chart comparing the number of marriages and divorces per
capita in the U.S. between 1900, 1950, and 2000.

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
SetA=pd.DataFrame(US.loc[(US['Year'] == 1900)])
SetB=pd.DataFrame(US.loc[(US['Year'] == 1950)])
SetC=pd.DataFrame(US.loc[(US['Year'] == 2000)])
final = pd.concat([SetA, SetB, SetC])
plt.rcParams["figure.figsize"] = (10, 10)
final.plot(x="Year", y=["Marriages_per_1000", "Divorces_per_1000"], kind="bar")
plt.xlabel('Year')
plt.ylabel('Per Capita')
plt.title('No. of Marriages and Divorces per capita in the U.S. between 1900, 1950, and 2000')
plt.show()
```

# Question 8

Create a horizontal bar chart that compares the deadliest actors in Hollywood. Sort
the actors by their kill count and label each bar with the corresponding actor's name.

```python
import pandas as pd
Hollywood = pd.read_csv("actor_kill_counts.csv")
Hollywood.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
Hollywood.sort_values(by=['Count'], inplace=True)
Actor = Hollywood['Actor']
Killcount = Hollywood['Count']
plt.rcParams["figure.figsize"] = (10,10)
plt.barh(Actor, Killcount, align='center', alpha=0.5)
plt.xlabel('Kill count')
plt.ylabel('Actors Name')
plt.title('Deadliest Actors in Hollywood')
plt.show()
```

# Question 9

Create a pie chart showing the fraction of all Roman Emperors that were
assassinated.

```python
import pandas as pd
Roman = pd.read_csv("roman-emperor-reigns.csv")
Roman.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
Roman = pd.read_csv('roman-emperor-reigns.csv')
Data=pd.DataFrame(Roman.loc[(Roman['Cause_of_Death'] == "Assassinated")])
my_labels = Data['Emperor']
fig, ax = plt.subplots()
ax.pie(Data["Length_of_Reign"],labels=my_labels,autopct='%1.1f%%',textprops={'fontsize': 15})
fig = plt.gcf()
fig.set_size_inches(20,20) 
plt.title('Assassinated Roman Emperors',fontsize=30)
plt.show()
```

# Question 10

Create a scatter plot showing the relationship between the total revenue earned by
arcades and the number of Computer Science PhDs awarded in the U.S. between
2000 and 2009.

```python
import pandas as pd
PHD = pd.read_csv("arcade-revenue-vs-cs-doctorates.csv")
PHD.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
PHD = pd.read_csv('arcade-revenue-vs-cs-doctorates.csv')
fig, ax = plt.subplots(figsize=(8, 8))
for i in range(len(PHD['Year'])):
    ax.scatter(PHD['Total Arcade Revenue (billions)'][i], PHD['Computer Science Doctorates Awarded (US)'][i])
ax.set_title('Total revenue earned by arcades and the number of Computer Science PhDs awarded in the U.S. between years 2000 and 2009')
ax.set_xlabel('Total Revenue (In Billion)')
ax.set_ylabel('No of Computer Science PhDs')
plt.show()
```
